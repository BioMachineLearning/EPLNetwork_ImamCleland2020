```python
import shutil
from pathlib import Path
import genData_extended, multiOdorTest_epl, multiOdorTest_hashtable, produce_figures
```

```python
## Settings

# Decide if you want to use the precompiled files and analysis, or recompile
COMPILE_PICKLE = False
COMPILE_EPL_ANALYSIS = True
COMPILE_HASHTABLE_ANALYSIS = True

# Set paths
dir_data = Path('data')
dir_pickle_files_compiled = Path('pickle_files_compiled')
dir_pickle_files = Path('pickle_files_current')
dir_results_compiled = Path('results_compiled')
dir_results = Path('results_current')
dir_figures = Path('results/figures')
dir_figures.mkdir(exist_ok=True, parents=True)
```

```python
# Creating pickle files from raw data

if COMPILE_PICKLE:
    dir_pickle_files.mkdir(exist_ok=True, parents=True)    
    genData_extended.run(dir_data, dir_pickle_files)
    
else:
    # pass
    # Check if the destination directory exists
    if dir_pickle_files.exists():
        # If it exists, remove it and its contents
        shutil.rmtree(dir_pickle_files)
    # Copy the pickle files to the destination directory
    shutil.copytree(dir_pickle_files_compiled, dir_pickle_files)
```

```python
# Running EPL analysis

if COMPILE_EPL_ANALYSIS:
    multiOdorTest_epl.run(dir_pickle_files, (dir_results / 'results_epl'))
else:
    # Check if the destination directory exists
    if (dir_results / 'results_epl').exists():
        # If it exists, remove it and its contents
        shutil.rmtree(dir_results / 'results_epl')
    # Copy the result files to the destination directory
    shutil.copytree(dir_results_compiled / 'results_epl', dir_results / 'results_epl')
```

```python
# Running Hashtable analysis

if COMPILE_HASHTABLE_ANALYSIS:
    multiOdorTest_hashtable.run(dir_pickle_files, (dir_results / 'results_hashtable'))
else:
    # Check if the destination directory exists
    if (dir_results / 'results_hashtable').exists():
        # If it exists, remove it and its contents
        shutil.rmtree(dir_results / 'results_hashtable')
    # Copy the result files to the destination directory
    shutil.copytree(dir_results_compiled / 'results_hashtable', dir_results / 'results_hashtable')
```

```python
# Producing figures

produce_figures.run(results_dir_parent=dir_results, figures_dir=dir_figures)
```
